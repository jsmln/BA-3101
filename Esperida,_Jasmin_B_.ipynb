**Importing Libraries**

```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
```

**Loading Dataset**

```python
productSales = pd.read_csv("product_sales.csv")
```

**First Look**

```python
# Rows and Columns

print(productSales.shape)
```

```python
# First 5 rows

print(productSales.head())
```

```python
# Last 5 rows

print(productSales.tail())
```

**Structure and Vitals**

```python
# Column Names and Datatypes

print(productSales.info())
```

```python
# Column Names

print(productSales.columns)
```

**Data Quality Check**

```python
# Missing Values

print("Missing Values per column: ")
print(productSales.isnull().sum())
```

```python
# Duplicate Rows

print("Number of Duplicate Rows: ")
print(productSales.duplicated().sum())
```

**Statistical Summary**

```python
# Numerical Columns Summary

print(productSales.describe())
```

```python
#Categorical Columns Summary

for col in productSales.select_dtypes(include=['object']):
    print(f"\nSummary for column: {col}")
    print(productSales[col].value_counts())
```

**Data Visualization**

```python
# Histogram of Numerical Data

plt.figure(figsize=(8,5))
plt.hist(productSales['years_as_customer'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Years as Customer')
plt.ylabel('Frequency')
plt.title('Distribution of Customer Loyalty (Years as Customer)')
plt.show()
```

```python
# Bar Chart for Categorical Columns

productSales['sales_method'].value_counts().plot(kind='bar')
plt.title("Distribution of Sales Methods")
plt.xlabel("Sales Method")
plt.ylabel("Count")
plt.show()
```

```python
# Scatter Plot (Revenue vs. Site Visits)

plt.scatter(productSales['nb_site_visits'], productSales['revenue'])
plt.title("Revenue vs Site Visits")
plt.xlabel("Number of Site Visits")
plt.ylabel("Revenue")
plt.show()
```

**Data Observation Analysis**

- The first look showed the dataset has multiple records with sales, customer, and revenue details.

- From the structure and vitals, we confirmed a mix of categorical (sales method, state) and numerical (units sold, revenue, years as customer, site visits) data.

- The data quality check revealed some missing values and potential duplicates, which need cleaning before deeper analysis.

- The statistical summary highlighted typical ranges: revenue and units sold mostly fall within mid-values but include outliers, while categorical columns like sales method and state show uneven distributions.

The visualizations reinforced these findings:

- Histograms showed most customers are new and revenue is concentrated at lower ranges with a few high-value outliers.

- Bar charts revealed that certain sales methods and states dominate.

- Scatter plots indicated positive relationships between units sold and revenue, as well as between site visits and revenue.

**Data Observation Synthesis**

The observations suggest that while businesses effectively acquires new customers and generates steady mid-range sales, there is reliance on specific sales channels and regions. Outliers in revenue and units sold highlight unique cases worth deeper investigation. To strengthen insights, data cleaning is essential, followed by strategies focused on customer retention, diversification of sales methods, and regional growth.
